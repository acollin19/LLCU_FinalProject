```python

# Genius API client authenticated using access token
genius = lyricsgenius.Genius('4HxFZukx7SgzqloFquyHEGBIaiLHc06mLAixH-ThU9MXKmQa1jeX0GJtRpzCPf2T', timeout=20)

# Retrieve the lyrics for given song title/artist
def get_lyrics(title, artist):
    song = genius.search_song(title, artist)
    if song is not None:
        return song.lyrics
    else:
        return "InstrumentalSong"

with open('lyrics1952.txt', 'w') as f:
    for title, artist in songs1952:
        lyrics = get_lyrics(title, artist)
        f.write(lyrics + '\n' + "New Song"+'\n' )
```
