```python
#imports
import pandas as pd
import re
```

```python
with open('lyrics2022.txt', 'r') as f:
    text = f.read()

text = re.sub(r'\s*\[.*?\]\s*', '', text) #removing brackets like [chorus], [verse] ...
splitBySong = text.split("New Song")

for song in splitBySong:
    lyrics_start = song.find("Lyrics")+len("Lyrics") # Removing gibbersish before song lyrics starts
    lyrics = song[lyrics_start:]
    lyrics = re.sub(r'(\d+)Embed', r'\1', lyrics)


    print(lyrics)
```
